```python
train["Fare"] = (train["Fare"] - train["Fare"].mean())/train["Fare"].std()
test["Fare"] = (test["Fare"] - test["Fare"].mean())/test["Fare"].std()

train["Fare"].fillna(train.groupby("Pclass")["Fare"].transform("mean"), inplace=True)
test["Fare"].fillna(test.groupby("Pclass")["Fare"].transform("mean"), inplace=True)
```

```python
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
test["FamilySize"] = test["SibSp"] + test["Parch"] + 1
```

```python
train["Age"] = (train["Age"] - train["Age"].mean())/train["Age"].std()
test["Age"] = (test["Age"] - test["Age"].mean())/test["Age"].std()

train["Pclass"] = (train["Pclass"] - train["Pclass"].mean())/train["Pclass"].std()
test["Pclass"] = (test["Pclass"] - test["Pclass"].mean())/test["Pclass"].std()

train["Sex"] = (train["Sex"] - train["Sex"].mean())/train["Sex"].std()
test["Sex"] = (test["Sex"] - test["Sex"].mean())/test["Sex"].std()

train["Emvbarked"] = (train["Emvbarked"] - train["Emvbarked"].mean())/train["Emvbarked"].std()
test["Emvbarked"] = (test["Emvbarked"] - test["Emvbarked"].mean())/test["Enmbarked"].std()

train["Title"] = (train["Title"] - train["Title"].mean())/train["Title"].std()
test["Title"] = (test["Title"] - test["Title"].mean())/test["Title"].std()

train["FamilySize"] = (train["FamilySize"] - train["FamilySize"].mean())/train["FamilySize"].std()
test["FamilySize"] = (test["FamilySize"] - test["FamilySize"].mean())/test["FamilySize"].std()
```

```python
features_drop = ['SibSp', 'Parch']
train = train.drop(features_drop, axis=1)
test = test.drop(features_drop, axis=1)
train = train.drop(['PassengerId'], axis=1)
```

```python
train_data = train.drop('Survived', axis=1)
target = train['Survived']

train_data.shape, target.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

import numpy as np
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
clf = SVC()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```
