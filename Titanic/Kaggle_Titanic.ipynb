```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
np.random.seed(7)

print('tensorflow version : ', tf.__version__)
print('keras version : ', keras.__version__)
with tf.device("/gpu:0"):
    np.random.seed(3)
    tf.random.set_seed(3)
```

```python
x_data = train.values[:, [1,2,3,4,5,6,7]]
y_data = train.values[:, [0]]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, 
                                                    test_size=0.2, random_state=7)
```

```python
X_train.shape, X_test.shape
```

```python
model = Sequential()
model.add(Dense(64, input_shape=(7,), activation='relu'))
Dropout(0.2)
model.add(Dense(32, activation='relu'))
Dropout(0.2)
model.add(Dense(1, activation='sigmoid'))
model.compile(loss="mse", optimizer="Adam", metrics=["accuracy"])
model.summary()
```

```python
epochs = 50
batch_size = 2
hist = model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size = batch_size, epochs=epochs)
```

```python
plt.figure(figsize=(18, 6))

plt.subplot(1, 2, 1)
plt.plot(hist.history["accuracy"], label="accuracy")
plt.plot(hist.history["val_accuracy"], label="val_accuracy")
plt.title("accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(hist.history["loss"], label="loss")
plt.plot(hist.history["val_loss"], label="val_loss")
plt.title("loss")
plt.legend()

plt.show()
```

```python
test_data = test.drop("PassengerId", axis=1).copy()

prediction = model.predict(test_data).reshape(test_data.shape[0])

prediction = np.where(prediction > 0.5, 1, 0)
```
