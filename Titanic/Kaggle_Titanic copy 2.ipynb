```python
train_test_set = [train, test]
```

```python
for dataset in train_test_set:
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
    dataset['First_Name'] = dataset['Name'].str.extract('([A-Za-z\-]+)\, ', expand=False)
    
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }

for dataset in train_test_set:
    dataset['Title'] = dataset['Title'].map(title_mapping)
```

```python
train.head()
```

```python
name_list = np.concatenate((train["First_Name"], test["First_Name"]), axis=0)
class_name = np.unique(name_list)
first_name = train["First_Name"]
group = []
for i in range(len(first_name)):
    for j in range(len(class_name)):
        if first_name[i] == class_name[j]:
            group.append(j)


train["group"] = np.array(group)
group = []

first_name = test["First_Name"]
for i in range(len(first_name)):
    for j in range(len(class_name)):
        if first_name[i] == class_name[j]:
            group.append(j)
test["group"] = np.array(group)
```

```python
train.drop('Ticket', axis=1, inplace=True)
test.drop('Ticket', axis=1, inplace=True)

train.drop('Cabin', axis=1, inplace=True)
test.drop('Cabin', axis=1, inplace=True)
```

```python
bins = [-1, 5, 10, 17, 30, 77, 1000]
label = [0, 1, 2, 3, 4, 5]
for dataset in train_test_set:
    dataset['Fare'] = pd.cut(dataset['Fare'], bins=bins, labels=label).cat.codes
```

```python


train["Fare"].fillna(train.groupby("Pclass")["Fare"].transform("mean"), inplace=True)
test["Fare"].fillna(test.groupby("Pclass")["Fare"].transform("mean"), inplace=True)

train["Fare"] = (train["Fare"] - train["Fare"].mean())/train["Fare"].std()
test["Fare"] = (test["Fare"] - test["Fare"].mean())/test["Fare"].std()
```

```python
import random
train["Embarked"].fillna(random.randint(0, 2), inplace=True)
test["Embarked"].fillna(random.randint(0, 2), inplace=True)
    
```

```python
train.info()
```

```python
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
test["FamilySize"] = test["SibSp"] + test["Parch"] + 1
```

```python
test["Age"] = (test["Age"] - train["Age"].mean())/train["Age"].std()
train["Age"] = (train["Age"] - train["Age"].mean())/train["Age"].std()

test["Pclass"] = (test["Pclass"] - train["Pclass"].mean())/train["Pclass"].std()
train["Pclass"] = (train["Pclass"] - train["Pclass"].mean())/train["Pclass"].std()

test["Sex"] = (test["Sex"] - train["Sex"].mean())/train["Sex"].std()
train["Sex"] = (train["Sex"] - train["Sex"].mean())/train["Sex"].std()

test["Embarked"] = (test["Embarked"] - train["Embarked"].mean())/train["Embarked"].std()
train["Embarked"] = (train["Embarked"] - train["Embarked"].mean())/train["Embarked"].std()

test["Title"] = (test["Title"] - train["Title"].mean())/train["Title"].std()
train["Title"] = (train["Title"] - train["Title"].mean())/train["Title"].std()

test["FamilySize"] = (test["FamilySize"] - train["FamilySize"].mean())/train["FamilySize"].std()
train["FamilySize"] = (train["FamilySize"] - train["FamilySize"].mean())/train["FamilySize"].std()

test["group"] = (test["group"] - train["group"].mean())/train["group"].std()
train["group"] = (train["group"] - train["group"].mean())/train["group"].std()
```

```python
features_drop = ['SibSp', 'Parch']
train = train.drop(features_drop, axis=1)
test = test.drop(features_drop, axis=1)
train = train.drop(['PassengerId'], axis=1)
```

```python
train_data = train.drop('Survived', axis=1)
target = train['Survived']
train_data = train_data.drop('First_Name', axis=1)
test_data = test.drop("First_Name", axis=1)

train_data.shape, target.shape, test_data.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

import numpy as np
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split

train_x, val_x, train_y, val_y = train_test_split(train_data, target, test_size=0.3, random_state=78)

train_x.shape, train_y.shape, val_x.shape, val_y.shape
```

```python
model = Sequential()
model.add(Flatten(input_shape=(8, )))
model.add(Dense(64, activation='relu'))
Dropout(0.5)
model.add(Dense(32, activation='relu'))
Dropout(0.5)
model.add(Dense(32, activation='relu'))
Dropout(0.5)
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
epochs = 10
batch_size = 2

history = model.fit(train_x, train_y, batch_size=batch_size, epochs=epochs, verbose=1, validation_data=(val_x, val_y))
```

```python
plt.figure(figsize=(18, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history["accuracy"], label="accuracy")
plt.plot(history.history["val_accuracy"], label="val_accuracy")
plt.title("accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history["loss"], label="loss")
plt.plot(history.history["val_loss"], label="val_loss")
plt.title("loss")
plt.legend()

plt.show()
```
