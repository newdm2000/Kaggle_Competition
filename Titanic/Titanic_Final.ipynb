<h3>Feature Engineering</h3>
<h4>drop data</h4>
<h5>passengerid, name, cabin, ticket</h5>

<h5>fill na</h5>
<h6>embarked filled random data</h6>
<h6>age filled median by name title(train set)</h6>
<h6>mapping name, embarked, sex</h6>

```python
train_test_set = [train_data, test_data]

drop_list = ["PassengerId", "Cabin", "Name", "Ticket", "SibSp", "Parch"]

title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }
sex_mapping = {"male": 0, "female": 1}
em_mapping = {"S":0, "C":1, "Q":2}

for dataset in train_test_set:
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Sex'] = dataset['Sex'].map(sex_mapping)
    dataset['Embarked'] = dataset['Embarked'].map(em_mapping)
    

for dataset in train_test_set:
    dataset['Embarked'].fillna(np.random.randint(0,2), inplace=True)
    dataset["Age"].fillna(train_data["Age"].groupby(train_data["Title"]).transform("median"), inplace=True)
    dataset["Fare"].fillna(train_data["Fare"].groupby(train_data["Embarked"]).transform("mean"), inplace=True)
    
bins = [-1, 5, 10, 17, 30, 77, 1000]
label = [0, 1, 2, 3, 4, 5]

for dataset in train_test_set:
    dataset['Fare'] = pd.cut(dataset['Fare'], bins=bins, labels=label).cat.codes\

    

for dataset in train_test_set:
    dataset["Pclass"] = dataset["Pclass"] - 1
    dataset["FamilySize"] = dataset["SibSp"] + dataset["Parch"]
    
bins = [-1, 11, 18, 22, 27, 33, 40, 66, 1000]
label = [0, 1, 2, 3, 4, 5, 6, 7]

for dataset in train_test_set:
    dataset['Age'] = pd.cut(dataset['Age'], bins=bins, labels=label).cat.codes

for item in drop_list: 
        train_data = train_data.drop(item, axis=1)
        test_data = test_data.drop(item, axis=1)
        
train_y = train_data["Survived"]
train_data = train_data.drop("Survived", axis=1)
```

```python
#RandomForest
param_grid1 = { 'n_estimators': [10, 20, 50],
    'min_samples_split': [5, 10],
    'min_samples_leaf': [5, 10],        
    'max_depth' : [None, 1, 3, 5],
    'criterion' :['gini', 'entropy']
}
#Ada 
param_grid2 = {'n_estimators': [10, 20, 50],
     'learning_rate' : [ 0.1, 0.2, 0.5]
}
#Extra Trees
param_grid3 = { 'n_estimators': [20, 50, 100],
    'min_samples_split': [5, 10],
    'min_samples_leaf': [5, 10],        
    'max_depth' : [None, 1, 3, 5],
    'criterion' :['gini', 'entropy']
}

#KN
param_grid4 = { 'n_neighbors': [10, 20, 50],
     'algorithm':['auto', 'ball_tre', 'kd_tree', 'brute'],
     'n_jobs' : [None, 1, 3, 5]
}  

param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
param_grid5 = {'C': param_range, 'gamma': param_range}
             


param_grid = [param_grid1,
              param_grid2,
              param_grid3,
              param_grid4,
              param_grid5
]

classifier_names = [RandomForestClassifier(),
                    AdaBoostClassifier(),
                    ExtraTreesClassifier(),
                    KNeighborsClassifier(),
                    SVC(random_state=1)
]                    

i=1
for param_gr,clf_name in zip(param_grid,classifier_names):
    print(i,clf_name)
    modeli = GridSearchCV(clf_name,param_gr )
    modeli.fit(x_train, y_train)
    print('The best parameters:',modeli.best_params_)
    i=i+1
```

```python
#RF_CL
model1 = RandomForestClassifier(criterion = 'gini', max_depth = None, min_samples_leaf = 5, min_samples_split = 5, n_estimators = 20)
model1.fit(x_train, y_train)


#AdaBoost
model2 = AdaBoostClassifier(learning_rate = 0.2, n_estimators = 50)
model2.fit(x_train, y_train)


#Extra Trees
model3 = ExtraTreesClassifier(criterion = 'gini', max_depth = None, min_samples_leaf = 5, min_samples_split = 10, n_estimators = 100 )
model3.fit(x_train, y_train)


#KNeighborsClassifier
model4 = KNeighborsClassifier(n_neighbors = 10 ,algorithm = 'auto' ,n_jobs = None)
model4.fit(x_train, y_train)

model5 = SVC(random_state=1, C=100.0, gamma= 0.01, probability=True)
model5.fit(x_train, y_train)

#ensemble method - votig, I will use setting - SOFT 
voting_model = VotingClassifier(estimators=[('RF', model1), ('ADA', model2), ('Extra', model3), ('KN', model4), ('SVC', model5) ], voting='soft')

voting_model.fit(x_train, y_train)

for voting in (model1, model2, model3, model4, model5, voting_model):
    voting.fit(x_train, y_train)
    y_pred = voting.predict(x_test);
    print('Accuracy score for the Voting approach is:',voting.__class__.__name__, accuracy_score(y_test, y_pred));
```

```python
#stacking

combined_datatest = np.stack((model1.predict_proba(x_train)[:,1],
                              model2.predict_proba(x_train)[:,1],
                              model3.predict_proba(x_train)[:,1],
                              model4.predict_proba(x_train)[:,1],
         ),
         axis=-1)

clf_blender = LogisticRegression( solver='lbfgs',multi_class='multinomial').fit(combined_datatest, y_train)


combined_test_datatest = np.stack((model1.predict_proba(x_test)[:,1],
                                   model2.predict_proba(x_test)[:,1],
                                   model3.predict_proba(x_test)[:,1],
                                   model4.predict_proba(x_test)[:,1],          
         ),
         axis=-1)


print("Stacking Approach Accuracy Score is: ", str(accuracy_score(clf_blender.predict(combined_test_datatest), y_test)))

#Learning curve after usiong the Stacking approach
model = LinearRegression()
```
